## 3. Frecuencia crítica

```python
#Convertir las columnas 'fecha' a tipo datetime para poder comparar fechas
SAP2['FECHA'] = pd.to_datetime(SAP2['FECHA'])
AM['FECHA'] = pd.to_datetime(AM['FECHA'])

resultados = []

#Variables para almacenar la fecha de inicio y la avería de inicio
fecha_inicio = AM['FECHA'].min()
averia_inicio = SAP2.iloc[0]['REPORTE AVERÍA']

#Iterar sobre cada fila del DataFrame SAP
for index, row in SAP2.iterrows():
    
    fecha_sap = row['FECHA']
    aviso = row['Aviso']
    averia = row['REPORTE AVERÍA']
    sistema = row['SISTEMA']
    diagnosticos_sap = row['DIAGNÓSTICOS'].split(', ')
    
    if averia == averia_inicio:
        
        #Obtener las fechas de AM que están dentro del rango de fecha_inicio y fecha_sap
        fechas_am = AM[(AM['FECHA'] == fecha_inicio) | (AM['FECHA'] >= fecha_inicio) & (AM['FECHA'] <= fecha_sap)]['FECHA'].tolist()
        defectos_am=AM[(AM['FECHA'] == fecha_inicio) | (AM['FECHA'] >= fecha_inicio) & (AM['FECHA'] <= fecha_sap)]['DEFECTO'].tolist()
        averia_inicio = averia
        fecha_inicio = fecha_sap
        
    else:
        
        #Actualizar la fecha de inicio y obtener las fechas de AM anteriores a fecha_sap
        fecha_inicio = AM['FECHA'].min()
        fechas_am = AM[(AM['FECHA'] == fecha_inicio) | (AM['FECHA'] >= fecha_inicio) & (AM['FECHA'] <= fecha_sap)]['FECHA'].tolist()
        defectos_am=AM[(AM['FECHA'] == fecha_inicio) | (AM['FECHA'] >= fecha_inicio) & (AM['FECHA'] <= fecha_sap)]['DEFECTO'].tolist()
        averia_inicio = averia
        fecha_inicio = fecha_sap

    
    #Contar las repeticiones de diagnósticos en los defectos encontrados
    conteo_repetidos = sum(1 for d in defectos_am for diag in diagnosticos_sap if diag in d)
    
    #Contar los defectos por cada diagnóstico
    conteo_por_diagnostico = {}
    for diagnostico in diagnosticos_sap:
        conteo = sum(1 for d in defectos_am if diagnostico in d)
        conteo_por_diagnostico[diagnostico] = conteo
    
    #Agregar los resultados a la lista de resultados
    resultados.append({'aviso': aviso, 'Avería': averia,'sistema': sistema ,'fecha': fecha_sap, 'diagnósticos': diagnosticos_sap, 'conteo_defectos': conteo_repetidos, 'conteo_por_diagnostico': conteo_por_diagnostico, 'defectos_am': defectos_am, 'fechas_am': fechas_am})

resultados_df = pd.DataFrame(resultados)
```

```python
resultados_df
```

Desagregamos cada fila con las respectivas fechas

```python
filas_diagnosticos = []

for index, row in resultados_df.iterrows():
    diagnosticos = row['diagnósticos']
    fechas_am = row['fechas_am']
    for diagnostico, fecha_am in zip(diagnosticos, fechas_am):
        nueva_fila = {
            'aviso': row['aviso'],
            'Avería': row['Avería'],
            'sistema': row['sistema'],
            'fecha': row['fecha'],
            'diagnóstico': diagnostico,
            'conteo_defectos': row['conteo_defectos'],
            'conteo_por_diagnostico': row['conteo_por_diagnostico'][diagnostico],
            'defectos_am': row['defectos_am'],
            'fecha_am': fecha_am
        }
        filas_diagnosticos.append(nueva_fila)

resultados_diagnosticos_df = pd.DataFrame(filas_diagnosticos)
```
